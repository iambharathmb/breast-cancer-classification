```python
# loding dataset from sklear
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
# adding Dataframes to dataset
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
#Save the dataset for later use
data.to_csv('beast-cancer-classification.csv', encoding='utf-8')
```

```python
# seprating the features and target
x = data.drop(columns='label')
y = data['label']
```

```python
x.shape , y.shape
```

```python
# standalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
# splitting the data into training and testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Building The Neaural Network

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of nural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30, )),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer = 'adam', 
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
              )
```

```python
# Training the neural network
history = model.fit(x_train, y_train, validation_split = 0.1, epochs = 10)
```

```python
# # Save the model in .h5 formate
# model.save("beast-cancer-model.h5")
```

```python
# visulizating training accuracy and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epochs")

plt.legend(['training data', 'validation data'])

plt.show()
```

```python
# visulizating training loss and Validation loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epochs")

plt.legend(['training loss', 'validation loss'])

plt.show()
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(x_test, y_test)
print("Accuracy:", accuracy)
print("Loss:", loss)
```
